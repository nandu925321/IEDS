```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#i#mport mpl_toolkits
%matplotlib inline
```

```python
# Read the data here
data = pd.read_csv("kc_house_data.csv")
```

```python
print(data.yr_renovated)
```

```python
# replace NaN with 0
data.fillna(value=0,inplace=True)
print(data.yr_renovated)
```

```python
# basic graph showing the number of bed room counts
data['bedrooms'].value_counts().plot(kind='bar')
plt.title('number of Bedroom')
plt.xlabel('Bedrooms')
plt.ylabel('Count')
plt.show()
```

```python
# basic graph showing prive versus living space
plt.scatter(data.price,data.sqft_living)
plt.title("Price vs Square Feet")
plt.show()
```

```python
print(data.date)
```

```python
# drop price from the train data

conv_dates = [values[0:4] for values in data.date ]
#print(conv_dates)
data['date'] = conv_dates
labels = data['price']
train1 = data.drop(['id', 'price'],axis=1)
# test train split
from sklearn.cross_validation import train_test_split
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.10,random_state =2)
print(x_test)
print(y_test)
```

```python
#Build the model - A simple linear regression model
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
y_pred = reg.predict(x_test)
reg.score(x_test,y_test)
```

```python
# calculate r squared error
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```

```python
# get the p value
from sklearn .feature_selection import f_regression
score = f_regression(x_train,y_train)
print(score[1])
```

```python
# Ridge regression
from sklearn.linear_model import Ridge
reg_r = Ridge()
reg_r.fit(x_train,y_train)
y_pred = reg_r.predict(x_test)
reg_r.score(x_test,y_test)
```

```python
# Lasso regression
from sklearn.linear_model import Lasso
reg_l= Lasso(max_iter=10000)
reg_l.fit(x_train,y_train)
y_pred = reg_l.predict(x_test)
reg_l.score(x_test,y_test)
```
